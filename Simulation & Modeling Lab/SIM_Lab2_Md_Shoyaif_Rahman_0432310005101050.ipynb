**2.Using NumPy, compute:**

● Shape of A

● Determinant

● Rank

● Eigenvalues

● Inverse (only if determinant ≠ 0)

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

A = np.array([[0+2,5+1],[2*0,5+2]])

#-------------shape of matrix A------------
shape = A.shape
print("shape of matrix a is: ",shape)

#---------------determinent of A--------
det_A = np.linalg.det(A)
print("determinent of A",det_A)

#rank of  matrix A
rank_A = np.linalg.matrix_rank(A)
print("rank of a matrix is : ",rank_A)


#eigen values
eigen_values,eigen_vectors = np.linalg.eig(A)
print("eigen values : ",eigen_values)


#find inverse of  matrix A
if np.linalg.det(A)!=0:
  inv_A = np.linalg.inv(A)
  print("inverse of matrix A is :\n",inv_A)
else: print("the determinent of the matrix is 0")

```

##  Comparison & Explanation

### 1. How did the determinant change and why?
The determinant increased from **14** to **21** because the first diagonal element changed from **2 to 3**.  
For triangular matrices, the determinant is the **product of diagonal elements**.

---

### 2. Did the rank change?
No.  
Both matrices have **rank 2** because their rows are linearly independent.  
Changing a single value does not affect linear independence.

---

### 3. How did the eigenvalues change?
Only one eigenvalue changed.

- A → eigenvalues: **2, 7**
- B → eigenvalues: **3, 7**

For triangular matrices, **eigenvalues are the diagonal elements**, so only the modified diagonal value affected the eigenvalues.

---

### 4. Is B easier or harder to invert than A?
Matrix **B is easier and more stable to invert**.

**Reason:**
- Larger determinant
- Further from being singular
- Better numerical stability
